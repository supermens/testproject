### 1.3 用户的地址判断
- 有的用户地址是在国外，需要提取出国内的用户地址

```python
# 函数判断地址是否是中文，即：用户地址是否是中国
# han = /^[\u4e00-\u9fa5]+$/
# 这个函数暂时不用
import re
def judge_han(ch):
    if re.match(r"^[\u4e00-\u9fa5]+$",ch):
        return ch
```

```python
# 将中国城市提取出来，因为有些用户的地址是在国外
address = []
for ch in list(user_unique_movie_data.user_address):
#     result = judge_han(ch)
    if re.match(r"^[\u4e00-\u9fa5]+$",ch):
        address.append(ch)
print(address)
```

```python
china_province = ['北京','天津','上海','重庆','河北','山西','辽宁','吉林','黑龙江','江苏','浙江','安徽','福建','江西','山东','河南','湖北','湖南','广东','海南','四川','贵州','云南','陕西','甘肃','青海','台湾','内蒙古','广西','西藏','宁夏','新疆','香港','澳门']
```

```python
langName_city = ['黑龙江','内蒙古']
zhixiashi = ['北京','天津','重庆','香港','澳门']
```

```python
# 用户省份的列表
user_province = []
for add in address:
    if add[:2] in china_province:
        user_province.append(add[:2])
    elif add[:3] in china_province:
        user_province.append(add[:3])
print(user_province)
```

```python
# 用户省份数据 [(省份，人数)]
province_data = []
for pro in set(user_province):
    province_data.append((pro,user_province.count(pro)))
province_data
```

```python
from pyecharts import Map
map = Map("豆瓣电影评分人分布图",width=1000,height=500)
attr,value = map.cast(province_data)
map.add(
    "",
    attr,
    value,
    maptype="china",
    is_visualmap=True,
    visual_text_color="#000",
)
```

### 3.1 计算每部电影的平均评分

```python
movie_star = movie_data[:].groupby("movie_name")["star"].mean()
# DataFrame(movie_star)
# DataFrame(movie_star.index)
DataFrame(movie_star.values,columns=["star",])
```

### 3.2 合并信息
- movie_name/movie_show/time/star

```python
movie_unique = movie_data.drop_duplicates(subset="movie_name").loc[:,("movie_name","movie_show","time")]
movie_unique.index = [i for i in range(len(movie_unique.index))]

movie_unique["star"]= DataFrame(movie_star.values,columns=["star",])
```

```python
movie_unique=movie_unique.round(1)
```

```python
# # 查看日期的个数
# movie_show_data = movie_unique.movie_show
# list(movie_show_data)
# x_axis = []
# for x in set(movie_show_data):
#     print(x)
```

```python
# type(movie_unique["movie_show"])
# 将电影中的上映时间转为时间格式
movie_unique["movie_show"] = pd.to_datetime(movie_unique['movie_show'])
# 将上映时间设置为索引
movie_unique = movie_unique.set_index("movie_show")
```

```python
print(movie_unique.head())
```

```python
movie_unique.index
```

```python
# 获取2018年到2019年的数据
movie_3D = movie_unique['2018':'2019']
movie_3D
```

```python
# 按照某种日期格式显示数据
# df_period = movie_3D.to_period('M')
# df_period
```

```python
# 轴标签
x_name = sorted(list(set(movie_3D.index)))
# x_name # 上映时间
y_name = sorted(list(set(movie_3D.time)))
# print(x_name,y_name)
x_name
```

```python
data_xyz = []
for i in range(len(movie_3D)):
    x = x_name.index(movie_3D.index[i])
    y = y_name.index(movie_3D.iloc[i,1])
    z = movie_3D.iloc[i,2]
    data_xyz.append([x,y,z])
data_xyz
```

```python
from pyecharts import Bar3D
bar3D=Bar3D("2018年到2019年各电影的信息","",
            title_pos="center",width=800,height=500)
bar3D.add("",x_name,y_name,data_xyz,is_label_show=True,is_visualmap=True,
          xaxis3d_name="上映时间",yaxis3d_name="电影时长",zaxis3d_name="电影评分",
         visual_range=[1,5],grid3d_width=150,grid3d_depth=50)
```
